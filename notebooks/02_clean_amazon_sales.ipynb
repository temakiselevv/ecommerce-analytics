```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/Amazon Sale Report.csv', sep=',', low_memory=False)
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_').str.replace('-', '_')

df = df.drop(columns=['index', 'unnamed:_22'])
df = df.dropna(subset=['sku', 'date'])
```

```python
df['fulfilment'] = df['fulfilment'].replace({
    'Merchant': 'Merchant Easy Ship'
})
df.drop(columns=['fulfilled_by'], inplace=True)
```

```python
df['date'] = pd.to_datetime(df['date'], format='%m-%d-%y', errors='coerce')

str_cols = ['order_id', 'style', 'sku', 'asin', 'ship_city', 'ship_state', 'ship_country', 'promotion_ids']
df[str_cols] = df[str_cols].astype('string')

cat_cols = ['category', 'status', 'size', 'courier_status', 'fulfilment', 'sales_channel', 'ship_service_level', 'currency']
df[cat_cols] = df[cat_cols].astype('category')
```

```python
df = df[df['qty'] > 0]
df = df[df['amount'] >= 0]
df = df.drop_duplicates()

df.info()
```

```python
df.to_csv('../data/cleaned/amazon_sales_clean.csv', index=False)
df.to_parquet('../data/cleaned/amazon_sales_clean.parquet', index=False)
```
