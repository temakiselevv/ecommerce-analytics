```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/raw/Sale Report.csv', sep=',', low_memory=False)
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')
df.rename(columns= {'sku_code' : 'sku', 'design_no.' : 'design_no'}, inplace=True)
```

```python
df.drop(columns=['index'], inplace=True)
df['sku'] = df['sku'].replace('#REF!', np.nan)
df.dropna(subset=['sku'], inplace=True)
df = df.dropna(subset=['stock'])
df = df[df['stock'] >= 0]
```

```python
str_cols = ['sku', 'design_no','color']
df[str_cols] = df[str_cols].astype('string')

cat_cols = ['category','size']
df[cat_cols] = df[cat_cols].astype('category')

df['stock'] = df['stock'].astype('int32')

for col in str_cols:
    df[col] = df[col].str.strip()
```

```python
df.to_csv('../data/cleaned/stock_clean.csv', index=False)
df.to_parquet('../data/cleaned/stock_clean.parquet', index=False)
```
